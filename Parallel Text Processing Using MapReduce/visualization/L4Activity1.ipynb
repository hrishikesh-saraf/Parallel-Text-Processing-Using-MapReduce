```r
#Loading required packages
library(wordcloud)
```

```r
#Reading from the output file of wordcount of hashtags
d<-read.table( "wordcount.txt",  sep = "|", fill=TRUE)
```

```r
#Displaying wordcloud with minimum frequency 1
wordcloud(words = d$V1, freq = d$V2, min.freq = 1,max.words=200, random.order=FALSE, rot.per=0.35,colors=brewer.pal(8, "Dark2"))
```
